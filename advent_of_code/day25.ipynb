```python
matrix = []
for line in open('input25.txt', 'r'):
    if line != '\n':
        line = line.lstrip().rstrip()
        matrix.append(list(line))


def pprint(mat):
    for x in mat:
        for y in x:
            print(y, end='')

        print()

def get_dest(char, i, j, N, M):
    if char == '>':
        return i, (j + 1) % M
    elif char == 'v':
        return (i + 1) % N, j
    else:
        return i, j


def is_vacant(matrr, pos):
    return matrr[pos[0]][pos[1]] == '.'

def countt(mat):
    x, y = 0, 0
    for i in range(len(mat)):
        for j in range(len(mat[0])):
            x += mat[i][j] == '>'
            y += mat[i][j] == 'v'

    return i, j


def move(new):
    N, M = len(new), len(new[0])
    mat = [[y for y in x] for x in new]

    # scan first to find out which > can move in this round
    indices = []
    for i in range(N):
        for j in range(M):
            if mat[i][j] == '>':
                destx, desty = get_dest(mat[i][j], i, j, N, M)
                if is_vacant(mat, (destx, desty)):
                    indices.append([destx, desty])
                else:
                    indices.append([i, j])
                
    for i in range(N):
        for j in range(M):
            if mat[i][j] == '>':
                mat[i][j] = '.'
    
    for pos in indices:
        mat[pos[0]][pos[1]] = '>'

    
    # then scan all v that can move
    indices = []
    for i in range(N):
        for j in range(M):
            if mat[i][j] == 'v':
                destx, desty = get_dest(mat[i][j], i, j, N, M)

                
                if is_vacant(mat, (destx, desty)):
                    indices.append([destx, desty])
                else:
                    indices.append([i, j])
                

    for i in range(N):
        for j in range(M):
            if mat[i][j] == 'v':
                mat[i][j] = '.'
    
    for pos in indices:
        mat[pos[0]][pos[1]] = 'v'

   

    return mat


    

count = 0
while True:
    
    next = move(matrix)
    count += 1
    
    print("we are at", count)
    
    if next == matrix:
        print(count)
        break
    matrix = [[x for x in y] for y in next]
    
```
